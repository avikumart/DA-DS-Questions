### **Q2c: Write a user defined function to generalize the task presented in Q2a to any URL that retrieves the content of the webpage. Your function should take a URL string as an input and return a correctly formulated BeautifulSoup instance as the output. In your function definition, ensure that appropriate exceptions are raised to the user (through status codes) if they pass in malformed/incorrect URLs. Write two test cases for your function - one with a working URL and another with an URL that gets a 404 response**
---

```python
def soup_generator(url):

  response = requests.get(url, headers=head)
  if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
  elif response.status_code == 404:
    raise Exception("Error: URL not found")
  else:
    print("Error: ", response.status_code)
  return soup
```

### **Q4a: Titles of all the movies on the page as a Python list**
---
---

```python
# write a use defined function of Titles of all the movies on the page as a Python
def get_titles(soup):
  titles = []
  for title in soup.find_all("div", class_="card style_1"):
    titles.append(title.h2.text)
  return titles

get_titles(soup)
```

### **Q4b: User ratings of all the movies on the page as a Python list**
---



```python
#write a function of User ratings of all the movies on the page as a Python list
def get_ratings(soup):
  ratings = []
  for rating in soup.find_all("div", class_="user_score_chart"):
    ratings.append(rating['data-percent'])
  return ratings
```

```python
get_ratings(soup)
```

### **Q4c: HTML content of all the individual pages of movies collected into a Python list**
---

```python
def get_html_content(soup):
  movie_urls = []
  for url in soup.find_all("div", class_="card style_1"):
    movie_urls.append(url.h2.a['href'])
  html_content = []
  for url in movie_urls:
    html_content.append(soup_generator("https://www.themoviedb.org" + url))
  return html_content
```

```python
html_content = get_html_content(soup)
```

### **Q4d: Genres of all the movies on the page as a Python list**
---

```python
import re

def get_genres(html_list):
  genres = []
  for soup in html_list:
    genre = re.sub(r'\xa0', "", soup.find("span", class_="genres").text.strip())
    genres.append(genre)
  return genres
```

```python
get_genres(html_content)
```

### **Q4e: Cast of all the movies on the page as a Python list**


---



```python
# Cast of all the movies on the page as a Python list
def cast_of_movies(html_list):
  cast_urls = []
  for soup in html_list:
    cast_urls.append(soup.find("p", class_="new_button").a['href'])
  cast_list = []
  for url in cast_urls:
    soup = soup_generator("https://www.themoviedb.org" + url)
    crew = soup.find("ol", class_="people credits")
    cast = []
    for member in crew.find_all("div", class_="info"):
      cast.append(member.a.text.strip())
    cast_list.append(cast)
  return cast_list
```

```python
cast_of_movies(html_content)
```

### **Q5: Write an user defined function that returns a pandas data frame with following data**

1. Titles of the movies listed on the page
2. User ratings of the movies listed on the page
3. Genres of the movies listed on the page
4. A cast of the movies listed on the page


---



---



```python
# Generate dataframe
import pandas as pd

def get_dataframe(soup, html_list):
  df = pd.DataFrame()
  df['Titles'] = get_titles(soup)
  df['User Ratings'] = get_ratings(soup)
  df['Genres'] = get_genres(html_list)
  df['Cast'] = cast_of_movies(html_list)
  return df
```

```python
get_dataframe(soup, html_content)
```
